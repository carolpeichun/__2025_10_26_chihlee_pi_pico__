```python
chinese = int(input("請輸入國文成績:"))
english = int(input("請輸入英文成績:"))
math = int(input("請輸入數學成績:"))

total = chinese + english + math    #程式原則上會以敘述式撰寫, 要讓敘述式有意義就要將它儲存(方便之後再取用), 在寫好的敘述式前面加"="並宣告變數(名稱), 即可把該敘述式丟進"="前面所宣告的變數, 之後就可以用該變數呼叫此敘述式。
print("總分:",total,"分",sep="")    #呼叫內建的function print(), 本身是敘述式, 會將敘述式後面的"內容"印出來, 如:print("總分:",total,"分",sep="")
                                   #print()裡, 引數sep分隔符號已預設" ", 若想改預設(另外給定其它值), 一律放最後, 比較建議的改法, 是"引數名稱=想給定的值", 如:sep=""。

average = round(total / 3,2)       #round(數值,小數位)
         #round(total / 3,ndigits=2) ndigits=2是比較清楚的寫法, "引數名稱=想給定的值", 但此內建函數的引數可省略引數名稱, 因說明書裡只有有設預設的只有一個引數, 即ndigits。
print("平均:",average,"分",sep="")
```
